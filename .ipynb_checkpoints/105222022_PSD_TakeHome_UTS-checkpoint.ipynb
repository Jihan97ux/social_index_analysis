# 3.Insights and Analysis and 4. Data Visualization


1. Identifikasi Potensial Area dengan menggunakan Weighted Sustainability Index
    Tujuan: Cari Provinsi yang memiliki potensial untuk dilakukannya pengembangan, misalnya:


*   Tingkat kemiskinan (POVERTY) rendah.
*   Non-electric households (NOELECTRIC) rendah.
*   Akses ke air bersih (TAPWATER) tinggi.
*   Tingkat pendidikan rendah (LOWEDU) rendah.
*   Persentase buta huruf (ILLITERATE) rendah



    Model Matematika :
     S = w1*(1-POVERTY) + w2*(1-NOELECTRIC) + w3*TAPWATER + w4*(1-LOWEDU) + w5(1-ILLITERATE)

*   variabel wi merupakan konstanta seberapa penting faktor-faktor diatas
*   Sustainability Index yang semakin besar menandakan Area tersebut semakin berpotensi untuk dikembangkan, dan semakin kecil Sustainability Index semakin  rentan Area tersebut dan memiliki tantangan yang besar terhadap pembangunan


     

```python
plt.figure(figsize=(14, 8))
sns.boxplot(data=df[['POVERTY', 'NOELECTRIC', 'TAPWATER', 'LOWEDU', 'ILLITERATE']])
plt.title("Persebaran Data yang akan digunakan untuk  Sustainability Index", fontsize=16)
plt.ylabel("Nilai", fontsize=12)
plt.show()
```

```python
columns_to_check = ['POVERTY', 'NOELECTRIC', 'TAPWATER', 'LOWEDU', 'ILLITERATE']

for col in columns_to_check:
      print(f"{col}: Min = {df[col].min()}, Max = {df[col].max()}")
```

```python
w1 = 0.3
w2 = 0.2
w3 = 0.2
w4 = 0.15
w5 = 0.15

def calculate_s(df):
    df['Sustainability_Index'] = (w1 * (1 - df['POVERTY']) +
               w2 * (1 - df['NOELECTRIC']) +
               w3 * (df['TAPWATER']) +
               w4 * (1 - df['LOWEDU']) +
               w5 * (1 - df['ILLITERATE']))
    return df

df = calculate_s(df)
df = df.sort_values(by='Sustainability_Index', ascending=False).reset_index(drop=True)

print("Distrik Potensial berdasarkan skor Sustainability Index:")
print(df[['PROVINCE', 'DISTRICTCODE', 'Sustainability_Index']])
```

```python
plt.figure(figsize=(14, 8))
sns.boxplot(data=df, x='PROVINCE', y='Sustainability_Index')
plt.title("Persebaran Sustainability Index berdasarkan Provinsi", fontsize=16)
plt.ylabel("Sustainability Index", fontsize=12)
plt.xlabel("Provinsi", fontsize=12)
plt.xticks(rotation=90)
plt.show()
```

```python
Sustainability_Index_perProvince = df.groupby('PROVINCE')['Sustainability_Index'].mean().reset_index()
```

### 6. Optional Tasks / Bonus Points
2. Identifikasi Potensi Demografi dengan Clustering

```python
df2 = df[['PROVINCE', 'CHILDREN', 'ELDERLY', 'FAMILYSIZE', 'POPULATION', 'GROWTH', 'FEMALE']].copy()

df2.loc[:, 'CHILDREN'] = df['CHILDREN'] * df['POPULATION']
df2.loc[:, 'ELDERLY'] = df['ELDERLY'] * df['POPULATION']
df2.loc[:, 'FEMALE'] = df['FEMALE'] * df['POPULATION']

df2 = df2.groupby('PROVINCE').agg({
    'FAMILYSIZE': 'mean',
    'POPULATION': 'sum',
    'GROWTH': 'mean',
    'CHILDREN': 'sum',
    'ELDERLY': 'sum',
    'FEMALE': 'sum'
}).reset_index()

df2['CHILDREN'] = df2['CHILDREN'] / df2['POPULATION']
df2['ELDERLY'] = df2['ELDERLY'] / df2['POPULATION']
df2['FEMALE'] = df2['FEMALE'] / df2['POPULATION']

df2.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

columns_to_cluster = ['CHILDREN', 'ELDERLY', 'FAMILYSIZE', 'POPULATION', 'GROWTH', 'FEMALE']

scaler = MinMaxScaler()
df_normalized = scaler.fit_transform(df2[columns_to_cluster])
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df_normalized)
    inertia.append(kmeans.inertia_)

plt.plot(range(1, 11), inertia, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
optimal_k = 3
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
df2['Cluster'] = kmeans.fit_predict(df_normalized)
```

```python
from sklearn.decomposition import PCA
import seaborn as sns

pca = PCA(n_components=2)
pca_result = pca.fit_transform(df_normalized)

df2['PCA1'] = pca_result[:, 0]
df2['PCA2'] = pca_result[:, 1]

sns.scatterplot(x='PCA1', y='PCA2', hue='Cluster', data=df2, palette='viridis')
plt.title('Clustering Result')
plt.show()
```

```python
print(df2.groupby('Cluster')[columns_to_cluster].mean())
```

Kesimpulan Cluster :


*   Cluster 0 : Mewakili Wilayah dengan populasi tinggi, dengan faktor lain medium
*   Cluster 1 : Mewakili Wilayah dengan pertumbuhan penduduk paling tinggi, angka jumlah famili yang paling besar, jumlah anak-anak terbanyak
*   Cluster 2 : Mewakili Wilayah dengan dominasi wanita terbanyak, tetapi memiliki angka pertumbuhan yang rendah serta dominasi lansia yang paling tinggi.
